# How Men and Women Commute in NYC

```python
import pandas as pd
```

```python
# load data in a dataframe ('df')
df = pd.read_csv('ACS_14_5YR_S0801_with_ann.csv')
```

```python
# preview the df
df.head()
```

```python
# remove margin of error columns
margins_of_error = [column for column in df.columns if 'MOE' in column]
df.drop(margins_of_error,inplace=True,axis=1)
```

```python
# preview df to make sure the MOE columns were successfully dropped
df.head()
```

```python
# save the first row because we will be dropping it
headers = df.columns
```

```python
# set the column headers equal to first row
df.columns = df.iloc[0]
```

```python
# drop the second row
df.drop(0,inplace=True,axis=0)
```

```python
# view male columns
[column for column in df.columns if 'MEANS OF TRANSPORTATION' in column and 'Male' in column]
```

```python
# view female columns
[column for column in df.columns if 'MEANS OF TRANSPORTATION' in column and 'Female' in column]
```

```python
# create male transportation df
male_columns = ['Id', 'Id2', 'Geography'] + [column for column in df.columns if \
                                             'MEANS OF TRANSPORTATION' in column and \
                                             'Male' in column and 'person carpool' not in column and \
                                             'Workers per car' not in column]
men = df[male_columns]

# create female transportation df
female_columns = ['Id', 'Id2', 'Geography'] + [column for column in df.columns if \
                                             'MEANS OF TRANSPORTATION' in column and \
                                             'Female' in column and 'person carpool' not in column and \
                                             'Workers per car' not in column]

women = df[female_columns]
```

```python
men
```

```python
# rename male columns
men = men.rename(columns={
        'Id': 'GeoID',
        'Id2': 'GeoID2',
        'Geography': 'GeoDisplay',
        'Male; Estimate; MEANS OF TRANSPORTATION TO WORK - Car, truck, or van': 'Drive',
        'Male; Estimate; MEANS OF TRANSPORTATION TO WORK - Car, truck, or van - Drove alone': 'Drive_alone',
        'Male; Estimate; MEANS OF TRANSPORTATION TO WORK - Car, truck, or van - Carpooled': 'Carpool',
        'Male; Estimate; MEANS OF TRANSPORTATION TO WORK - Public transportation (excluding taxicab)': 'Transit',
        'Male; Estimate; MEANS OF TRANSPORTATION TO WORK - Walked': 'Walk',
        'Male; Estimate; MEANS OF TRANSPORTATION TO WORK - Bicycle': 'Bike',
        'Male; Estimate; MEANS OF TRANSPORTATION TO WORK - Taxicab, motorcycle, or other means': 'Other',
        'Male; Estimate; MEANS OF TRANSPORTATION TO WORK - Worked at home': 'Work_from_home'
    })


# rename women columns
women = women.rename(columns={
        'Id': 'GeoID',
        'Id2': 'GeoID2',
        'Geography': 'GeoDisplay',
        'Female; Estimate; MEANS OF TRANSPORTATION TO WORK - Car, truck, or van': 'Drive',
        'Female; Estimate; MEANS OF TRANSPORTATION TO WORK - Car, truck, or van - Drove alone': 'Drive_alone',
        'Female; Estimate; MEANS OF TRANSPORTATION TO WORK - Car, truck, or van - Carpooled': 'Carpool',
        'Female; Estimate; MEANS OF TRANSPORTATION TO WORK - Public transportation (excluding taxicab)': 'Transit',
        'Female; Estimate; MEANS OF TRANSPORTATION TO WORK - Walked': 'Walk',
        'Female; Estimate; MEANS OF TRANSPORTATION TO WORK - Bicycle': 'Bike',
        'Female; Estimate; MEANS OF TRANSPORTATION TO WORK - Taxicab, motorcycle, or other means': 'Other',
        'Female; Estimate; MEANS OF TRANSPORTATION TO WORK - Worked at home': 'Work_from_home'
    })
```

```python
# preview men
men.head()
```

```python
# preview women
women.head()
```

```python
men
```

```python
# remove rows with dashes
men = men[men.Drive != '-']
women = women[women.Drive != '-']
```

```python
men.head()
```

```python
# Convert men data types to floats
men.Drive = men.Drive.astype(float)
men.Drive_alone = men.Drive_alone.astype(float)
men.Carpool = men.Carpool.astype(float)
men.Transit = men.Transit.astype(float)
men.Walk = men.Walk.astype(float)
men.Bike = men.Bike.astype(float)
men.Other = men.Other.astype(float)
men.Work_from_home = men.Work_from_home.astype(float)
```

```python
men.dtypes
```

```python
# Convert women data types to floats
women.Drive = women.Drive.astype(float)
women.Drive_alone = women.Drive_alone.astype(float)
women.Carpool = women.Carpool.astype(float)
women.Transit = women.Transit.astype(float)
women.Walk = women.Walk.astype(float)
women.Bike = women.Bike.astype(float)
women.Other = women.Other.astype(float)
women.Work_from_home = women.Work_from_home.astype(float)
```

```python
women.dtypes
```

```python
# save dfs to csv
men.to_csv('men.csv', index=False)
women.to_csv('women.csv', index=False)
```
